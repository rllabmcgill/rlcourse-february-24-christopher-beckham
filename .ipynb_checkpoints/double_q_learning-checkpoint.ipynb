Let us define the 'max bias' environment that Sutton presented in his book. To recall, we start in state *A*, and we can either go into the right terminal state (which gives us no reward), or we can go left into state *B*. If we are in state *B*, we can choose to go left again and hit the terminal state, but the reward we get is stochastic: on average we would get -0.1 reward, but the variance is also high (=1) so there is a good chance we can get a positive reward (which is better than a zero reward).

```python
Image("./max_bias_mdp.png", width=400)
```

Here is an OpenAI Gym implementation of this simple MDP.

```python
class MaxBiasEnv(discrete.DiscreteEnv):
    metadata = {'render.modes':['human','ansi']}
    def __init__(self):
        self.shape = (4,)
        nS = np.prod(self.shape)
        nA = 2
        # calculate transition probabilities
        self.LEFT = 0
        self.RIGHT = 1
        P = {}
        # HACK: reward is always zero here, but it's not entirely true.
        # we have to hack it in step() function
        for s in range(nS):
            P[s] = {a : [] for a in range(nA) }
            # observation, new_state, reward, is_done
            P[s][self.LEFT] = [(1.0, max(0, s-1), 0, True if max(0, s-1) == 0 else False)]
            P[s][self.RIGHT] = [(1.0, min(nS-1, s+1), 0, True if min(nS-1,s+1) == (nS-1) else False)]
        
        # we always start in state A (which is 2)
        isd = np.zeros(nS)
        isd[2] = 1.0
        
        super(MaxBiasEnv, self).__init__(nS, nA, P, isd)
        
    def _step(self, a):
        transitions = self.P[self.s][a]
        p, s, r, d = transitions[0]
        self.s = s
        self.lastaction=a
        # if we went from B to left terminal
        if self.s == 0:
            r = np.random.normal(-0.1, 1, 1)[0]
        return (s, r, d, {"prob" : p})
    
    def _render(self, mode='human', close=False):
        if close:
            return
        outfile = StringIO() if mode == 'ansi' else sys.stdout
        st = ['T', 'B', 'A', 'T']
        st[self.s] = '(' + st[self.s] + ')'
        outfile.write(" ".join(st))
        outfile.write("\n")
```

Let's test this environment to see if it works correctly. The game state that we print (via `env.render()`) prints out the four letters corresponding to the four states of the MDP, e.g. `T B A T`, where `T` corresponds to terminal. 

```python
env = MaxBiasEnv()
env.render()
print env.step(0)
env.render()
print env.step(0)
env.render()
```

(The parentheses around the letter show us the state we are currently in)

```python
env = MaxBiasEnv()
env.render()
print env.step(1)
env.render()
```

If we do 1000 iterations of the game where we always move into the left state and record the rewards and average them, we should get something close to the mean -0.1

```python
rs = []
for num_iter in range(1000):
    env = MaxBiasEnv()
    env.step(0)
    _,r,d,_ = env.step(0)
    rs.append(r)
print np.mean(rs)
```

Here we implement the Q-learning algorithm:

```python
Image("./qlearning.png", width=800)
```

```python
def Q_to_eps_greedy_policy(Q, eps=0.1):
    num_actions = Q.shape[1]
    pi = np.zeros(Q.shape[0],)
    for i in range(0, Q.shape[0]):
        if np.random.uniform(0,1) < eps:
            # choose a random action
            pi[i] = np.random.randint(0, num_actions)
        else:
            # take the argmax of Q[s,.]
            pi[i] = np.argmax(Q[i])
    return pi
```

```python
def q_learning(env, num_iters, eps, alpha, gamma=1.0):
    """
    :env:
    :num_iters: number of episodes
    :eps: epsilon
    :win_state:
    """
    Q = np.zeros((env.nS, env.nA))
    # repeat (for each episode that wins)
    reward_sums = []
    eps_lengths = []
    action_freqs = {a:[] for a in range(env.nA)}
    for num_iter in range(num_iters):
        action_freq = {a:0 for a in range(env.nA)}
        eps_length = 0
        reward_sum = 0
        env.reset()
        # initialise S
        S = env.s
        # repeat, for each step of episode
        while True:
            # choose A from S using policy derived from Q (e.g. eps-greedy)
            pi = Q_to_eps_greedy_policy(Q)
            A = pi[S]
            action_freq[A] += 1
            # take action A, observe R, S'
            _, R, is_done, _ = env.step(A)
            reward_sum += R
            eps_length += 1
            S_prime = env.s
            # we want to find argmax_a Q[S_prime, a]
            Q_as = []
            for a in range(env.nA):
                Q_as.append(Q[S_prime,a])
            A_max = np.argmax(Q_as)
            Q[S,A] = Q[S,A] + alpha*(R + (gamma*Q[S_prime,A_max]) - Q[S,A])
            S = S_prime
            eps_length += 1
            if is_done:
                break
        #print "eps length:", eps_length
        reward_sums.append(reward_sum)
        eps_lengths.append(eps_length)
        eps_length = 1.0 * sum([action_freq[key] for key in action_freq.keys()])
        for key in action_freqs:
            action_freqs[key].append( action_freq[key] / eps_length )
    return reward_sums, eps_lengths, action_freqs, Q
```

In particular, we are interested in seeing how the agent reacts in this environment, and more specifically, how often they take the left action. Statistically, we would expect the agent to always want to go right from the start state, since in the long run we will get zero reward (as opposed to the other terminal state, which we expect us to give -0.1 reward in the long run).

```python
def avg_arr(arr):
    return np.sum(np.asarray(arr),axis=0) / len(arr)
```

Here, we record the following and average these over 100 runs:
* The reward obtained as a function of the episode #
* The episode lengths as a function of the episode #
* The number of times we take the left action as a function of the episode #

We also get the Q table at the end of each of these runs and average this too.

We do these experiments with no discounting factor, an $\epsilon = 0.1$ and learning rate $\alpha = 0.1$

```python
arr_rewards = []
arr_lengths = []
arr_lefts = []
Q_avg = np.zeros((MaxBiasEnv().nS, MaxBiasEnv().nA))
N = 100
for j in range(N):
    print j,
    env = MaxBiasEnv()
    rewards, lengths, action_freqs, Q = q_learning(env, num_iters=300, eps=0.1, alpha=0.1)
    arr_rewards.append(rewards)
    arr_lengths.append(lengths)
    arr_lefts.append(action_freqs[0])
    Q_avg += Q
arr_rewards = avg_arr(arr_rewards)
arr_lengths = avg_arr(arr_lengths)
arr_lefts = avg_arr(arr_lefts)
Q_avg = Q_avg / N
```

```python
plt.plot(arr_lefts)
plt.xlabel("episode #")
plt.ylabel("freq of LEFT")
```

This is odd: between 50% and 60% of the time the left action is being taken.
Let's look at the Q table:

```python
print "Q[A] =", Q_avg[2]
print "Q[B] =", Q_avg[1]
```

This is also unusual. If we act greedy wrt this Q function, when we start in state *A*, we move left into state *B* (since this has the higher Q value), but when we're in state *B* we want to move back into state *A* since that also has the higher Q value. That effectively means that without $\epsilon$-greedy exploration, we're effectively stuck in an infinite loop. This also explains why the episode lengths are very long:

```python
plt.plot(arr_lengths)
```

Let's also look at the reward we get:

```python
plt.plot(arr_rewards)
```

```python
def dq_learning(env, num_iters, eps, alpha, gamma=1.0):
    """
    :env:
    :num_iters: number of episodes
    :eps: epsilon
    :win_state:
    """
    Q1 = np.zeros((env.nS, env.nA))
    Q2 = np.zeros((env.nS, env.nA))
    # repeat (for each episode that wins)
    reward_sums = []
    eps_lengths = []
    action_freqs = {a:[] for a in range(env.nA)}
    for num_iter in range(num_iters):
        action_freq = {a:0 for a in range(env.nA)}
        eps_length = 0
        reward_sum = 0
        env.reset()
        # initialise S
        S = env.s
        # repeat, for each step of episode
        while True:
            # choose A from S using policy derived from Q1+Q2 (e.g. eps-greedy)
            pi = Q_to_eps_greedy_policy(Q1+Q2)
            A = pi[S]
            action_freq[A] += 1
            # take action A, observe R, S'
            _, R, is_done, _ = env.step(A)
            reward_sum += R
            eps_length += 1
            S_prime = env.s
            # with 0.5 probability
            if np.random.randint(0,2) == 0:
                Q_as = [ Q1[S_prime,a] for a in range(env.nA) ]
                A_max = np.argmax(Q_as)
                Q1[S,A] = Q1[S,A] + alpha*(R + (gamma*Q2[S_prime,A_max]) - Q1[S,A])
            else:
                Q_as = [ Q2[S_prime,a] for a in range(env.nA) ]
                A_max = np.argmax(Q_as)
                Q2[S,A] = Q2[S,A] + alpha*(R + (gamma*Q1[S_prime,A_max]) - Q2[S,A])
            S = S_prime
            eps_length += 1
            if is_done:
                break
        #print "eps length:", eps_length
        reward_sums.append(reward_sum)
        eps_lengths.append(eps_length)
        eps_length = 1.0 * sum([action_freq[key] for key in action_freq.keys()])
        for key in action_freqs:
            action_freqs[key].append( action_freq[key] / eps_length )
    return reward_sums, eps_lengths, action_freqs, Q1+Q2
```

Let us now run 100 iterations of double Q-learning, where each iteration lasts 300 episodes.

```python
arr_rewards = []
arr_lengths = []
arr_lefts = []
Q_avg = np.zeros((MaxBiasEnv().nS, MaxBiasEnv().nA))
N = 100
for j in range(N):
    print j,
    env = MaxBiasEnv()
    rewards, lengths, action_freqs, Q = dq_learning(env, num_iters=300, eps=0.1, alpha=0.1)
    arr_rewards.append(rewards)
    arr_lengths.append(lengths)
    arr_lefts.append(action_freqs[0])
    Q_avg += Q
arr_rewards = avg_arr(arr_rewards)
arr_lengths = avg_arr(arr_lengths)
arr_lefts = avg_arr(arr_lefts)
Q_avg = Q_avg / N
```

We do the same thing as before, and now plot the frequency of left actions as a function of episode #:

```python
plt.plot(arr_lefts)
plt.xlabel("episode #")
plt.ylabel("freq of LEFT")
```

This is cool! We can see that the agent has learned to not take the left action.

We also expect our average reward to be zero:

```python
plt.plot(arr_rewards)
```

What is the problem with Q-learning in this case then? The problem is that, in the case of noisy samples, this inequality holds:

$\mathbb{E}\Big[ \max_{a} Q_{t}(s', a) \Big] > \max_{a} \mathbb{E}\Big[ Q_{t}(s', a) \Big]$

We can make this more clear with a specific example. Suppose we have a 2-armed bandit, both arms ($a_{0}$ and $a_{1}$) draw rewards from $N(0,1)$. Since both are drawn from a Gaussian distribution with mean 0, $\mathbb{E}\Big[ Q(s, a_{0}) \Big] = 0$ and $\mathbb{E}\Big[ Q(s, a_{1}) \Big] = 0$. Therefore, $\max_{b} \mathbb{E}\Big[ Q_{t}(s, b) \Big] = 0$ also!

However, the opposite, $\mathbb{E}\Big[ \max_{a} Q_{t}(s,a) \Big]$, will be much larger, and this is effectively what we are doing since this is our update equation in the Q-learning algorithm:

$Q(S,A) \leftarrow Q(S,A) + \alpha \Big[ R + \gamma \ max_{a} Q(S',a) - Q(S,A) \Big] $

This means that we can have a tendency to assign high values to state-action pairs when they really shouldn't be.

In the double Q-learning algorithm, we maintain two Q functions, $Q_{1}$ and $Q_{2}$. We can assign one of the Q's, say, $Q_{1}$, to be the function we use to choose the maximising action, and the other function $Q_{2}$ to be one we choose to estimate the value of a state-action pair. Then, our update for $Q_{1}$ becomes the following:

$Q_{1}(S,A) \leftarrow Q_{1}(S,A) + \alpha \Big[ R + \gamma \ Q_{2}(S',argmax_{a}Q_{1}(S',a)) - Q_{1}(S,A) \Big] $

The idea then becomes this: instead of one noisy Q function, we have two. When we're choosing an action to take in a state, we do it using the policy derived from a combination of $Q_{1}$ and $Q_{2}$, e.g. $\frac{Q_{1} + Q_{2}}{2}$. It is hoped that the noise from one of these Q's offsets the noise from the other $Q$. Therefore, we can see that the use of two Q's can help mitigate the maximisation bias problem.
